Dealing With Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['CREDIT_SCORE']=df['CREDIT_SCORE'].fillna(df['CREDIT_SCORE'].median())
df['ANNUAL_MILEAGE']=df['ANNUAL_MILEAGE'].fillna(df['ANNUAL_MILEAGE'].median())
```

Dealing With Outliers

```python
! pip install datasist
```

```python
df.info()
```

```python
from datasist.structdata import detect_outliers
outliers = detect_outliers(df,0,['CREDIT_SCORE','POSTAL_CODE','ANNUAL_MILEAGE','SPEEDING_VIOLATIONS','DUIS','PAST_ACCIDENTS'])
len(outliers)
```

```python
outliers2=detect_outliers(df,0,df.describe().columns)
len(outliers2)
```

```python
col=df.describe().columns.tolist()
col
```

```python
for i in col:
    outliers2=detect_outliers(df,0,[i])
    df[i].loc[outliers2]=df[i].median()
```

split input data and output data

```python
X = df.drop('OUTCOME',axis = 1)
y = df['OUTCOME']
```

```python
X
```

handle categorical data

```python
X = pd.get_dummies(X,drop_first = True)
```

```python
X
```

split data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(X,y, test_size = 0.2,random_state=22)
```

feature scaling

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns = scaler.get_feature_names_out())
x_train_scaled
```

```python
x_test_scaled = pd.DataFrame(x_test_scaled,columns = scaler.get_feature_names_out())
x_test_scaled
```
